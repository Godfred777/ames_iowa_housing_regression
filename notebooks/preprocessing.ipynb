```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# import seaborn as sns
```

```python
data = pd.read_csv("../data/train.csv")
data.head()
```

```python
missing_columns = data.columns[data.isnull().any()]
print("Columns with missing values:", missing_columns)
print(len(missing_columns))
```

```python
data.fillna({"Lot Frontage": 0}, inplace=True)
data.fillna({"Mas Vnr Type": "None"}, inplace=True)
data.fillna({"Mas Vnr Area": 0}, inplace=True)
data.fillna({"Garage Type": "None"}, inplace=True)
data.fillna({"Garage Yr Blt": 0}, inplace=True)
data.fillna({"Garage Finish": "None"}, inplace=True)
data.fillna({"Garage Cars": 0}, inplace=True)
data.fillna({"Garage Area": 0}, inplace=True)
data.fillna({"Garage Qual": "None"}, inplace=True)
data.fillna({"Garage Cond": "None"}, inplace=True)
data.fillna({"Bsmt Qual": "None"}, inplace=True)
data.fillna({"Bsmt Cond": "None"}, inplace=True)
data.fillna({"Bsmt Exposure": "None"}, inplace=True)
data.fillna({"BsmtFin Type 1": "None"}, inplace=True)
data.fillna({"BsmtFin Type 2": "None"}, inplace=True)
data.fillna({"Electrical": "None"}, inplace=True)
data.fillna({"Fireplace Qu": "None"}, inplace=True)
data.fillna({"Alley": "None"}, inplace=True)
data.fillna({"Pool QC": "None"}, inplace=True)
data.fillna({"Fence": "None"}, inplace=True)
data.fillna({"Misc Feature": "None"}, inplace=True)
```

```python
data.fillna({"BsmtFin SF 1": 0}, inplace=True)
data.fillna({"BsmtFin SF 2": 0}, inplace=True)
data.fillna({"Bsmt Unf SF": 0}, inplace=True)
data.fillna({"Total Bsmt SF": 0}, inplace=True)
data.fillna({"Bsmt Full Bath": 0}, inplace=True)
data.fillna({"Bsmt Half Bath": 0}, inplace=True)
```

```python
def mapping_ordinal_data(column):
    mapping = {
        "Ex": 5,  # Excellent
        "Gd": 4,  # Good
        "TA": 3,  # Average/Typical
        "Fa": 2,  # Fair
        "Po": 1  , # Poor
    }
    
    data[column] = data[column].map(mapping)
    return data.head()
    
```

```python
mapping_ordinal_data("Exter Qual")
mapping_ordinal_data("Exter Cond")
mapping_ordinal_data("Bsmt Qual")
mapping_ordinal_data("Bsmt Cond")
mapping_ordinal_data("Heating QC")
mapping_ordinal_data("Kitchen Qual")
mapping_ordinal_data("Fireplace Qu")
mapping_ordinal_data("Garage Qual")
mapping_ordinal_data("Garage Cond")
```

```python
data.fillna({"Fireplace Qu": 0}, inplace=True)
```

```python
categorical_columns = data.select_dtypes(include=['object', 'str']).columns
categorical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_categorical_data = encoder.fit_transform(data[categorical_columns])

for i, column in enumerate(encoder.categories_):
    print(f"Features {i}: {encoder.categories_[i]}")
```

```python
print(encoded_categorical_data.shape)
print(data.shape)
print(encoded_categorical_data)
```

```python
# Get feature names for the encoded data
encoded_feature_names = encoder.get_feature_names_out(categorical_columns)

# Create a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_categorical_data, columns=encoded_feature_names, index=data.index)

# Drop the original categorical columns from data
data = data.drop(columns=categorical_columns)

# Concatenate the encoded DataFrame to the original data
data = pd.concat([data, encoded_df], axis=1)

data.head()
```

```python
data.to_csv("../data/preprocessed_data.csv", index=False)
```
